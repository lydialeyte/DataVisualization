```python
# Import necessary libraries
import pandas as pd
import numpy as np
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler

# Step 1: Load the CSV Data


```

```python
# Load the newly uploaded file to examine its structure and content
file_path = 'clustered_GM.csv'

# Read the file into a pandas DataFrame
df = pd.read_csv(file_path)

# Display the first few rows and column names to understand its structure
df.head(), df.columns
```

```python
# Step 1: Drop unnecessary columns and compute year-to-year differences
years = ['lbm_2002', 'lbm_2008', 'lbm_2012', 'lbm_2014', 'lbm_2016', 'lbm_2018', 'lbm_2020', 'lbm_2022']
df_trends = df[['ID'] + years].copy()

# Compute year-on-year differences
for i in range(len(years) - 1):
    df_trends[f'diff_{years[i+1]}_{years[i]}'] = df_trends[years[i+1]] - df_trends[years[i]]

# Only keep difference columns for trend clustering
trend_columns = [col for col in df_trends.columns if col.startswith('diff_')]

# Standardize the differences
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
df_trends[trend_columns] = scaler.fit_transform(df_trends[trend_columns])

# Step 2: Apply clustering (e.g., KMeans)
from sklearn.cluster import KMeans

# Choose a number of clusters (e.g., 5)
n_clusters = 5
kmeans = KMeans(n_clusters=n_clusters, random_state=42)
df_trends['cluster'] = kmeans.fit_predict(df_trends[trend_columns])

'''# Step 3: Aggregate and visualize trends for each cluster
cluster_trends = df_trends.groupby('cluster')[trend_columns].mean()

# Reconstruct average year-to-year trends for each cluster
cluster_trends_cumulative = cluster_trends.cumsum(axis=1)
cluster_trends_cumulative.insert(0, 'Base', 0)
x_labels = ['2002-2008', '2008-2012', '2012-2014', '2014-2016', '2016-2018', '2018-2020', '2020-2022']
# Plot the trends for each cluster
plt.figure(figsize=(12, 8))
for cluster in cluster_trends_cumulative.index:
    trend = cluster_trends_cumulative.loc[cluster].values
    plt.plot(range(len(trend)), trend, label=f'Cluster {cluster}')

plt.title('Clustered Trends in Livability Scores')
plt.xlabel('Time Steps (Differences Between Consecutive Years)')
plt.ylabel('Cumulative Trend (Standardized Differences)')
plt.legend(title="Cluster")

plt.xticks(range(len(x_labels) + 1), ['Base'] + x_labels, rotation=45)  # Adjust x-ticks to include 'Base'
plt.grid()
plt.show()'''
# Step 3: Compute normalized lbm scores for each cluster
'''
map each cluster's average normalized year-to-year differences back to the original lbm scores for the respective years.
Use the year-to-year differences to compute the normalized lbm scores cumulatively for each cluster.
'''
x_labels = ['2002', '2008', '2012', '2014', '2016', '2018', '2020', '2022']
# Normalize the original lbm scores
df_trends[years] = scaler.fit_transform(df_trends[years])

# Compute average normalized lbm for each cluster
cluster_means = df_trends.groupby('cluster')[years].mean()

# Step 4: Visualize normalized lbm scores over years
plt.figure(figsize=(12, 8))
for cluster in cluster_means.index:
    plt.plot(years, cluster_means.loc[cluster], label=f'Cluster {cluster}')

plt.title('Normalized LBM Scores Over Time by Cluster')
plt.xlabel('Year')
plt.xticks(range(len(x_labels)), x_labels, rotation=45)  # Adjust x-ticks to include 'Base'
plt.ylabel('Normalized LBM Score')
plt.legend(title="Cluster")
plt.grid()

plt.show()
```

```python
cluster_means
cluster_means = cluster_means.rename(columns={'lbm_2002':'2002',  'lbm_2008':'2008',  'lbm_2012':'2012',  'lbm_2014':'2014', 'lbm_2016':'2016', 'lbm_2018':'2018',  'lbm_2020':'2020', 'lbm_2022':'2022'})
```

```python
output_df = df_trends[['ID', 'lbm_2002',  'lbm_2008',  'lbm_2012',  'lbm_2014', 'lbm_2016', 'lbm_2018',  'lbm_2020', 'lbm_2022', 'cluster']].copy()
for i, label in enumerate(x_labels[:]):  
    output_df[label] = cluster_means.iloc[:, i].values[df_trends['cluster']]

# Save to CSV
output_file_path = "clustered_GM_2.csv"
output_df.to_csv(output_file_path, index=False)

output_file_path
```
